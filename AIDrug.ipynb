```python
from torch_geometric.nn import GINConv, global_mean_pool
import torch.nn.functional as F
import torch.nn as nn
import torch

class GIN(torch.nn.Module):
    def __init__(self, num_tasks=12):
        super().__init__()

        # MLP inside GIN block
        nn1 = nn.Sequential(nn.Linear(1, 64), nn.ReLU(), nn.Linear(64, 64))
        self.conv1 = GINConv(nn1)

        nn2 = nn.Sequential(nn.Linear(64, 64), nn.ReLU(), nn.Linear(64, 64))
        self.conv2 = GINConv(nn2)

        self.fc = nn.Linear(64, num_tasks)

    def forward(self, data):
        x, edge_index, batch = data.x.float(), data.edge_index, data.batch

        x = self.conv1(x, edge_index)
        x = F.relu(x)

        x = self.conv2(x, edge_index)
        x = F.relu(x)

        x = global_mean_pool(x, batch)  # graph-level pooling

        return self.fc(x)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = GIN().to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.BCEWithLogitsLoss()
```

✅ Next Step: Train the Improved GIN Edge-Aware Model

Below is the final training setup that will give you much higher ROC-AUC than before.

⭐ 1. Define the improved GIN model (GINEConv)

```python
from torch_geometric.nn import GINEConv, global_mean_pool
import torch.nn as nn
import torch.nn.functional as F
import torch

class GIN_edge(torch.nn.Module):
    def __init__(self, num_tasks=12):
        super().__init__()

        # Input node feature size = 7 (we defined 7 atom features)
        nn1 = nn.Sequential(nn.Linear(7, 64), nn.ReLU(), nn.Linear(64, 64))
        self.conv1 = GINEConv(nn1, edge_dim=4)

        nn2 = nn.Sequential(nn.Linear(64, 64), nn.ReLU(), nn.Linear(64, 64))
        self.conv2 = GINEConv(nn2, edge_dim=4)

        self.fc = nn.Linear(64, num_tasks)

    def forward(self, data):
        x = data.x
        edge_index = data.edge_index
        edge_attr = data.edge_attr
        batch = data.batch

        x = F.relu(self.conv1(x, edge_index, edge_attr))
        x = F.relu(self.conv2(x, edge_index, edge_attr))

        x = global_mean_pool(x, batch)
        return self.fc(x)
```

 we will now add Explainable AI (XAI) to your trained GIN model using Gradient-based Atom Attribution.

This code will generate atom-colored toxicity visualizations, with:

🔴 Red = atoms increasing toxicity

🟡 Yellow = mildly toxic

🔵 / ⚪ Neutral = low or no toxicity contribution

This works directly with your trained GIN-edge model.

```python
#✅ STEP 1 — Atom Attribution Function (Gradients)

import torch
from rdkit import Chem
from rdkit.Chem import Draw
import numpy as np

def explain_molecule(smiles, model, device, task_id=0):
    """
    Computes atom-level importance scores for a given molecule.
    Uses gradient magnitudes on node embeddings.
    """
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError("Invalid SMILES")

    # Build graph
    mol_obj = Chem.MolFromSmiles(smiles)
    g = mol_to_graph_full(mol_obj, np.zeros(12))   # dummy label
    g = g.to(device)

    model.eval()
    for param in model.parameters():
        param.requires_grad = True

    # Forward pass — keep gradient for node embeddings
    g.x.requires_grad_(True)
    out = model(g)
    logit = out[0, task_id]  # 1 score for the chosen toxicity task

    # Backward pass
    model.zero_grad()
    logit.backward()

    # Importance = gradient magnitude for each atom
    atom_importance = g.x.grad.abs().sum(dim=1).detach().cpu().numpy()

    # Normalize 0–1 for coloring
    atom_importance = atom_importance / (atom_importance.max() + 1e-6)

    return mol, atom_importance
```

```python
#✅ STEP 2 — RDKit Coloring Function

from rdkit.Chem.Draw import rdMolDraw2D
from rdkit.Geometry import Point2D

def draw_molecule_with_importance(mol, importance, size=(400, 300)):
    """
    Colors atoms based on importance scores using RDKit drawing.
    """
    # RDKit highlight color dictionary: atom_idx → (R,G,B)
    atom_colors = {}

    for i, score in enumerate(importance):
        # Convert score to color (red = toxic)
        r = float(score)
        g = float(1 - score)
        b = 0.2

        atom_colors[i] = (r, g, b)

    drawer = rdMolDraw2D.MolDraw2DSVG(size[0], size[1])
    drawer.drawOptions().addAtomIndices = True

    rdMolDraw2D.PrepareAndDrawMolecule(
        drawer, mol,
        highlightAtoms=list(range(len(importance))),
        highlightAtomColors=atom_colors
    )
    drawer.FinishDrawing()

    svg = drawer.GetDrawingText()
    return svg
```

```python
def explain_and_show(smiles, task_id=0):
    mol, importance = explain_molecule(smiles, model, device, task_id)

    svg = draw_molecule_with_importance(mol, importance)
    display(SVG(svg))

    # Print top toxic atoms
    print("Top contributing atoms (index: score):")
    for i in importance.argsort()[::-1][:5]:
        print(f"Atom {i}: {importance[i]:.3f}")
```

```python
from torch_geometric.data import Data
from rdkit import Chem

def mol_to_graph_full(mol):
    atoms = mol.GetAtoms()
    bonds = mol.GetBonds()

    # Node features
    x = torch.stack([atom_features(a) for a in atoms])

    edge_index = []
    edge_attr = []

    for b in bonds:
        i = b.GetBeginAtomIdx()
        j = b.GetEndAtomIdx()

        bf = bond_features(b)

        # forward
        edge_index.append([i, j])
        edge_attr.append(bf)

        # backward
        edge_index.append([j, i])
        edge_attr.append(bf)

    edge_index = torch.tensor(edge_index, dtype=torch.long).t()
    edge_attr = torch.stack(edge_attr)

    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr)

    return data
```

```python
def explain_molecule(smiles, model, device, task_id):
    model.eval()

    mol = Chem.MolFromSmiles(smiles)
    g = mol_to_graph_full(mol).to(device)

    g.x.requires_grad = True

    out = model(g)             # shape = [1, 12]
    logit = out[0, task_id]    # make it a scalar
    logit.backward()

    importance = g.x.grad.abs().sum(dim=1).cpu().numpy()

    return mol, importance
```

```python
from rdkit.Chem import Draw

from IPython.display import SVG

def visualize_importance(mol, importance):
    norm = (importance - importance.min()) / (importance.ptp() + 1e-8)

    # RDKit wants atom color mapping as {idx: (r,g,b)}
    atom_colors = {i: (float(w), 0.0, 0.0) for i, w in enumerate(norm)}

    drawer = Draw.MolDraw2DSVG(400, 400)
    drawer.drawOptions().useBWAtomPalette()

    drawer.DrawMolecule(
        mol,
        highlightAtoms=list(atom_colors.keys()),
        highlightAtomColors=atom_colors
    )

    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return SVG(svg)

```

```python
import streamlit as st
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GINEConv, global_mean_pool
from rdkit import Chem
from rdkit.Chem import rdMolDraw2D
import numpy as np
import io

# -----------------------------
# 1. Model + featurization
# -----------------------------

tox21_tasks = [
    "NR-AR", "NR-AR-LBD", "NR-AhR", "NR-Aromatase",
    "NR-ER", "NR-ER-LBD", "NR-PPAR-gamma",
    "SR-ARE", "SR-ATAD5", "SR-HSE", "SR-MMP", "SR-p53"
]

def atom_features(atom):
    """Must match what you used during training (7 dims)."""
    return torch.tensor([
        atom.GetAtomicNum(),               # 1
        atom.GetTotalDegree(),             # 2
        atom.GetFormalCharge(),            # 3
        atom.GetTotalNumHs(),              # 4
        int(atom.GetIsAromatic()),         # 5
        atom.GetImplicitValence(),         # 6
        int(atom.GetHybridization()),      # 7
    ], dtype=torch.float)

def bond_features(bond):
    """4-dim edge features (must match training)."""
    return torch.tensor([
        bond.GetBondTypeAsDouble(),        # 1
        int(bond.GetIsConjugated()),       # 2
        int(bond.GetStereo()),             # 3
        int(bond.IsInRing()),              # 4
    ], dtype=torch.float)

def mol_to_graph(mol):
    if mol is None:
        return None

    x = torch.stack([atom_features(a) for a in mol.GetAtoms()])

    edge_index = []
    edge_attr = []
    for b in mol.GetBonds():
        i = b.GetBeginAtomIdx()
        j = b.GetEndAtomIdx()
        f = bond_features(b)

        edge_index.append([i, j])
        edge_attr.append(f)

        edge_index.append([j, i])
        edge_attr.append(f)

    if len(edge_index) == 0:
        return None

    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    edge_attr = torch.stack(edge_attr)

    return Data(x=x, edge_index=edge_index, edge_attr=edge_attr)


class GIN_edge(torch.nn.Module):
    def __init__(self, num_tasks=12):
        super().__init__()

        nn1 = nn.Sequential(
            nn.Linear(7, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
        )
        self.conv1 = GINEConv(nn1, edge_dim=4)

        nn2 = nn.Sequential(
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
        )
        self.conv2 = GINEConv(nn2, edge_dim=4)

        self.fc = nn.Linear(64, num_tasks)

    def forward(self, data):
        x, edge_index, edge_attr, batch = (
            data.x, data.edge_index, data.edge_attr, data.batch
        )
        x = F.relu(self.conv1(x, edge_index, edge_attr))
        x = F.relu(self.conv2(x, edge_index, edge_attr))
        x = global_mean_pool(x, batch)
        return self.fc(x)


@st.cache_resource
def load_model(device="cpu"):
    model = GIN_edge().to(device)
    state_dict = torch.load("tox21_gnn.pt", map_location=device)
    model.load_state_dict(state_dict)
    model.eval()
    return model


def predict_smiles(smiles, model, device="cpu"):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError("Invalid SMILES string.")

    g = mol_to_graph(mol)
    if g is None:
        raise ValueError("Molecule has no bonds / graph could not be built.")

    # Create a batch of size 1
    from torch_geometric.data import Batch
    batch = Batch.from_data_list([g]).to(device)

    with torch.no_grad():
        logits = model(batch)
        probs = torch.sigmoid(logits).cpu().numpy()[0]

    return mol, probs


def explain_molecule(smiles, model, device="cpu", task_id=0):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError("Invalid SMILES string.")

    g = mol_to_graph(mol)
    if g is None:
        raise ValueError("Molecule has no bonds / graph could not be built.")

    from torch_geometric.data import Batch
    batch = Batch.from_data_list([g]).to(device)

    # We need gradients w.r.t. node features
    batch.x.requires_grad_(True)

    out = model(batch)           # shape [1, 12]
    logit = out[0, task_id]      # scalar for this task
    model.zero_grad()
    logit.backward()

    # importance per atom = gradient magnitude
    grad = batch.x.grad.detach().cpu()
    importance = grad.abs().sum(dim=1).numpy()    # [num_atoms]

    # normalize 0-1
    importance = (importance - importance.min()) / (importance.ptp() + 1e-8)
    return mol, importance


def draw_svg_molecule(mol, importance):
    """Return SVG string with atoms colored by importance."""
    atom_colors = {
        i: (float(w), 0.0, 0.0)  # red intensity
        for i, w in enumerate(importance)
    }

    drawer = rdMolDraw2D.MolDraw2DSVG(400, 300)
    drawer.drawOptions().useBWAtomPalette()
    rdMolDraw2D.PrepareAndDrawMolecule(
        drawer,
        mol,
        highlightAtoms=list(atom_colors.keys()),
        highlightAtomColors=atom_colors,
    )
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return svg


# -----------------------------
# 2. Streamlit UI
# -----------------------------

def main():
    st.set_page_config(page_title="Tox21 GNN Explainer", layout="wide")
    st.title("🧪 Molecular Toxicity Predictor (Tox21 GNN + XAI)")

    st.markdown(
        """
        Enter a SMILES string to predict **12 Tox21 toxicity endpoints** and see
        which atoms contribute most (red = higher contribution).
        """
    )

    device = "cpu"  # Vercel will be CPU-only
    model = load_model(device=device)

    col_input, col_options = st.columns([2, 1])

    with col_input:
        smiles = st.text_input(
            "SMILES",
            value="CC(=O)OC1=CC=CC=C1C(=O)O",  # Aspirin
            help="Example: CC(=O)OC1=CC=CC=C1C(=O)O",
        )

    with col_options:
        explain_task_name = st.selectbox(
            "Endpoint to explain (for atom heatmap):",
            tox21_tasks,
            index=0,
        )
        task_id = tox21_tasks.index(explain_task_name)

    if st.button("Run prediction"):
        if not smiles.strip():
            st.error("Please enter a SMILES string.")
            return

        try:
            mol, probs = predict_smiles(smiles, model, device=device)
        except Exception as e:
            st.error(f"Error: {e}")
            return

        # Show predictions
        st.subheader("Tox21 Toxicity Predictions")
        st.write(
            "Probability that the molecule is **active** in each assay "
            "(1 = toxic / active, 0 = inactive)."
        )

        pred_table = {
            "Endpoint": tox21_tasks,
            "Probability": [float(f"{p:.3f}") for p in probs],
        }
        st.table(pred_table)

        # XAI heatmap
        st.subheader(f"Atom-level explanation for: **{explain_task_name}**")
        try:
            mol_ex, importance = explain_molecule(
                smiles, model, device=device, task_id=task_id
            )
            svg = draw_svg_molecule(mol_ex, importance)

            import streamlit.components.v1 as components
            components.html(svg, height=320)

        except Exception as e:
            st.warning(f"Could not generate explanation: {e}")


if __name__ == "__main__":
    main()
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
from torch_geometric.loader import DataLoader

def evaluate_model(model, loader):
    model.eval()

    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)
            logits = model(batch)
            preds = torch.sigmoid(logits).cpu().numpy()
            labels = batch.y.cpu().numpy()

            all_preds.append(preds)
            all_labels.append(labels)

    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)

    results = {}

    for i, task in enumerate([
        "NR-AR", "NR-AR-LBD", "NR-AhR", "NR-Aromatase",
        "NR-ER", "NR-ER-LBD", "NR-PPAR-gamma",
        "SR-ARE", "SR-ATAD5", "SR-HSE", "SR-MMP", "SR-p53"
    ]):
        y_true = all_labels[:, i]
        y_pred = all_preds[:, i]

        # Skip tasks with no positives
        if len(np.unique(y_true)) < 2:
            continue

        roc = roc_auc_score(y_true, y_pred)

        # Binary predictions
        y_bin = (y_pred > 0.5).astype(int)
        acc = accuracy_score(y_true, y_bin)
        f1 = f1_score(y_true, y_bin)

        results[task] = {
            "ROC-AUC": roc,
            "Accuracy": acc,
            "F1-score": f1
        }

    return results
```
